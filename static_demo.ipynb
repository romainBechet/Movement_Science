```python
import pandas as pd 
from static import Static 
```

```python
data = pd.read_csv('data/data_static.txt', names = ['time', 'x', 'y'], delimiter='\t')
data
```

```python
## Create an instance of `Static`
trial_one = Static(data.x, data.y, 100)
```

```python
## Compute all parameters with one function: 
# Results can be stored into a variable. 

# As default, the ellipse CI is calculated with a confidence = 0.95. It can be changed by specifying the desired value 
trial_one.get_parameters(0.7)
```

```python
# Draw the ellipse with the desired CI. 
# 
trial_one.draw_ellipse_confidence_interval(
    conf =.90, # The default value for CI is 0.95 too and can be changed by passing the parameter into the function. 
    path_style = ':', # Style for the cop
    path_color = 'red', # Color for the cop
    ls = '--', # **kwargs from matpotlib.2DLines can be added for the ellipse style. 
    lw = 2, 
    color = 'blue' )
```

```python
# If we call the function again, the `area_cop` value is updated
trial_one.get_parameters()
```
